# Regime-Adaptive Intraday Alpha Model  
### Quantitative Research Framework, Universe Governance & Execution Protocol

---

## 1. Research Objective

This project develops and evaluates a systematic intraday alpha model operating on 5-minute bars.

The objective is to determine whether structural VWAP positioning, short-horizon momentum transitions, volatility conditioning, and disciplined risk management jointly produce statistically robust excess returns across a correlation-defined equity universe.

The emphasis of this research is not parameter optimization, but structural hypothesis validation, robustness verification, and bias-controlled execution modeling.

---

## 2. Structural Hypothesis

The core hypothesis is:

1. Persistent positioning relative to VWAP reflects intraday institutional directional bias.
2. Short-horizon EMA cross events represent momentum inflection points with decaying predictive power.
3. Volatility regimes influence signal reliability and trade expectancy.
4. Structural alignment (VWAP + trend + momentum) reduces false positives.
5. Anchored risk management improves convex payoff structure.
6. Intraday signal strength varies by time-of-day and should be statistically derived rather than assumed.

This framework defines the class of strategy under investigation.

---

## 3. Structural (Non-Optimized) Definitions

The following components define the modeling space and will **not** be statistically derived:

- 5-minute bar resolution  
- EMA(9,20) as short-horizon momentum transition pair  
- HMA(20) as structural trend filter  
- RSI(14) and default MACD parameters  
- Fixed notional position sizing  
- Initial capital baseline  

These parameters define the signal family and hypothesis space.  
They are not tuned to maximize performance.

---

## 4. Correlation-Defined Universe Construction

### 4.1 Anchor Asset

NVDA is designated as the structural anchor asset.

### 4.2 Universe Formation Methodology

For each candidate large-cap, high-liquidity U.S. equity:

\[
\rho_i = Corr(r_i, r_{NVDA})
\]

Where:
- \( r_i \) = daily log returns of stock i  
- \( r_{NVDA} \) = daily log returns of NVDA  
- Correlation window = 252 trading days  

Stocks are ranked by rolling correlation during a predefined universe formation period.

The top N correlated stocks form the trading universe.

### 4.3 Look-Ahead Prevention

- Universe is constructed using a calibration period.
- Universe is frozen prior to the main backtest window.
- No future information is used in selection.

This prevents survivorship and selection bias.

## 5. Statistically Governed Parameters

The following parameters will be empirically derived using controlled statistical analysis:

- ATR lookback length  
- ATR percentile threshold  
- EMA cross freshness window  
- VWAP structural activation delay  
- VWAP proximity band at cross  
- Stop-loss distance (VWAP-anchored)  
- Breakeven trigger distance  
- EMA/HMA reversal buffer  
- VWAP slope lookback windows  
- Intraday session participation windows  

No parameter will be selected based solely on peak Sharpe.

---

## 6. Session Derivation (Data-Driven Time Governance)

Intraday trading windows are not arbitrarily defined.

Instead:

1. Identify all EMA cross events.
2. Compute forward returns (k-bar horizon) after each cross.
3. Group events by time-of-day buckets.
4. Compute:
   - Mean forward return
   - t-statistic
   - Cross-to-cross move distribution

Time regimes where:
- Forward returns are statistically insignificant, or
- Cross-to-cross moves are economically negligible,

are excluded.

Only time windows with statistically credible signal strength are retained.

This replaces intuition with conditional expectation analysis.

---

## 7. Parameter Derivation Methodology

### A. Distributional Risk Analysis
- MAE distribution → stop placement
- MFE distribution → breakeven logic

### B. Alpha Decay Measurement
- Forward returns vs EMA cross age → freshness window
- Returns vs VWAP distance → proximity band

### C. Volatility Conditioning
- Strategy performance vs ATR percentile sweep
- Select stable plateau, not local maximum

### D. Local Robustness Testing
For each parameter:
- Evaluate narrow neighborhood
- Identify performance stability region
- Select midpoint within plateau

Parameters are selected for stability, not optimality.

---

## 8. Execution Model

### 8.1 Signal Timing

Signals are computed at bar close \( t \).

If entry conditions are satisfied at bar \( t \):

- Entry occurs at Open of bar \( t+1 \).

No intrabar assumptions are made.

### 8.2 Long/Short Symmetry

- Long PnL: \( (Exit - Entry) \times Qty \)
- Short PnL: \( (Entry - Exit) \times Qty \)

Performance is evaluated separately for:
- Long-only
- Short-only
- Combined

### 8.3 Trade Frequency

- Multiple trades per day allowed.
- One active position per asset at a time.
- No overlapping directional exposure per asset.

---

## 9. Risk Management Framework

- Stops anchored to VWAP at entry.
- Breakeven advancement upon favorable excursion.
- Structural invalidation exits (EMA/HMA reversal, MACD inflection).
- End-of-session flattening.
- End-of-data flattening.

All exits triggered at close of bar \( t \) execute at Open of bar \( t+1 \).

This enforces symmetrical fairness.

---

## 10. EMA Stability Verification

Although EMA(9,20) defines the structural hypothesis, robustness is verified by testing adjacent EMA pairs:

Examples:
- (8,21)
- (9,20)
- (10,22)

The objective is not optimization, but to confirm:

- Sharpe remains stable across nearby configurations.
- Performance does not collapse outside the baseline.
- Edge is structural rather than numerically fragile.

---

## 11. Robustness & Anti-Overfitting Protocol

Validation includes:

- Parameter neighborhood stability testing
- Walk-forward validation
- Out-of-sample testing
- Regime segmentation (volatility regimes)
- Cross-sectional validation across correlated assets
- EMA pair sensitivity analysis

The strategy must demonstrate stability across time, regime, and parameter neighborhoods.

---

## 12. Evaluation Metrics

Performance will be evaluated using:

- Sharpe ratio
- t-statistic of trade returns
- Maximum drawdown
- Win rate
- Trade count stability
- Regime-conditioned returns
- Return on deployed capital
- Long vs short performance decomposition

Statistical credibility and robustness are required for alpha validation.

---

## 13. Research Classification

This project represents:

> A correlation-defined, volatility-conditioned, regime-aware intraday alpha research framework with formal parameter governance and execution discipline.

The objective is not to produce an optimized backtest, but to determine whether a structurally defined signal family exhibits stable, statistically defensible alpha under controlled and bias-aware conditions.
