# Regime-Adaptive Intraday Alpha Model  
### Quantitative Research Framework, Parameter Governance & Validation Protocol

---

## 1. Research Objective

This project develops and evaluates a systematic intraday alpha model operating on 5-minute bars.

The objective is to determine whether structural VWAP positioning, short-horizon momentum transitions, volatility conditioning, statistically derived session participation, and disciplined risk management jointly produce robust excess returns across a correlation-defined equity universe.

The emphasis of this research is not performance maximization, but structural hypothesis validation, parameter stability, and regime robustness.

---

## 2. Structural Hypothesis

The core hypothesis is:

1. Persistent positioning relative to VWAP reflects intraday institutional directional bias.
2. Short-horizon EMA cross events represent momentum inflection points with decaying predictive value.
3. Volatility regimes materially affect signal reliability.
4. Structural alignment (VWAP + trend + momentum) reduces false positives.
5. Risk anchored to structural invalidation improves payoff convexity.
6. Intraday signal strength varies by time-of-day and must be statistically derived.

This framework defines the class of strategy under investigation.

---

## 3. Structural (Non-Optimized) Definitions

The following elements define the modeling space and will **not** be statistically tuned:

- 5-minute bar resolution  
- EMA(9,20) as baseline short-horizon momentum pair  
- HMA(20) as structural trend filter  
- RSI(14) and default MACD parameters  
- Fixed notional position sizing  
- Initial capital baseline  

These define the signal family and hypothesis space.

---

## 4. Correlation-Defined Universe Construction

### 4.1 Anchor Asset

NVDA is designated as the anchor asset.

### 4.2 Universe Selection Methodology

For each candidate equity:

$$
\rho_i = \mathrm{Corr}(r_i, r_{NVDA})
$$

Where:

- $r_i$ = daily log returns of stock $i$
- $r_{NVDA}$ = daily log returns of NVDA
- Correlation window = 252 trading days

Stocks are ranked by rolling correlation during a predefined universe formation period.

The top N correlated stocks form the trading universe.

The universe is frozen before the primary evaluation period to prevent look-ahead bias.

---

## 5. Statistically Governed Parameters

The following variables will be empirically derived using controlled statistical analysis:

### Volatility Conditioning
- ATR lookback length
- ATR percentile threshold for participation

### Momentum Freshness
- EMA cross freshness window (maximum bar age)

### Structural Activation
- VWAP structural activation delay (bars after two-sided confirmation)
- VWAP proximity band at cross event

### Risk Management
- Initial stop-loss distance (VWAP-anchored)
- Breakeven trigger distance
- EMA/HMA structural reversal buffer

### Structural Trend Filters
- VWAP slope lookback window (longs)
- VWAP slope lookback window (shorts)

### Intraday Participation
- Statistically derived session windows based on time-of-day signal significance

Each parameter will be selected based on:

- Distributional analysis (MAE/MFE)
- Conditional forward return studies
- Stability plateaus across neighborhoods
- Regime robustness
- Cross-sectional consistency

No parameter will be selected solely on peak Sharpe.

---

## 6. Session Derivation Protocol

Intraday participation windows are derived via:

1. Identify all EMA cross events.
2. Compute forward returns at multiple horizons.
3. Group events by time-of-day buckets.
4. Compute:
   - Mean forward return
   - t-statistic
   - Cross-to-cross move distribution
5. Exclude time regimes where:
   - Forward returns are statistically insignificant, or
   - Cross-to-cross moves are economically negligible.

Session participation is therefore data-driven, not arbitrary.

---

## 7. Execution Model

- Signals computed at close of bar $t$.
- Entry executed at Open of bar $t+1$.
- All exits triggered at close of bar $t$ execute at Open of bar $t+1$.

PnL:

Long:
$$
(Exit - Entry) \times Qty
$$

Short:
$$
(Entry - Exit) \times Qty
$$

Performance evaluated separately for:
- Long-only
- Short-only
- Combined

One active position per asset at a time.
Multiple trades per day permitted.
No overnight exposure.

## 8. Time Governance & Validation Framework

Data Range: 2016–2026

Two complementary validation approaches will be conducted.

---

### A. Static Parameter Validation (Fixed Governance Test)

**Calibration Period:** 2016–2019  

Used exclusively to:

- Derive statistically governed parameters  
- Conduct neighborhood robustness testing  
- Estimate regime-conditioned stability  

Parameters are locked after this phase.

**Out-of-Sample Evaluation:** 2020–2026  

No recalibration permitted.

This tests structural permanence of the derived parameter set.

---

### B. Rolling Recalibration Validation (Adaptive Test)

To simulate realistic deployment under regime drift:

Rolling structure:

- Train 2016–2019 → Test 2020  
- Train 2017–2020 → Test 2021  
- Train 2018–2021 → Test 2022  
- Train 2019–2022 → Test 2023  
- Train 2020–2023 → Test 2024  
- Train 2021–2024 → Test 2025  

Each test year is strictly out-of-sample relative to its training window.

This evaluates regime adaptability and parameter stability under rolling recalibration.

---

### Final Holdout

2026 remains fully untouched until final evaluation.

Represents simulated live deployment.

## 9. Structural Stability & Perturbation Analysis

Although EMA(9,20) defines the baseline structural hypothesis, robustness will be evaluated across neighboring parameter configurations to ensure the strategy captures a genuine structural dynamic rather than numerical coincidence.

### A. EMA Pair Stability

Test neighboring EMA pairs:

- (8,21)  
- (9,20)  
- (10,22)  

Objective:

- Confirm Sharpe stability across adjacent configurations  
- Ensure performance does not collapse outside baseline  
- Validate that the edge is momentum-structural rather than parameter-fragile  

---

### B. VWAP Proximity Band Perturbation

The VWAP deviation threshold used for cross qualification will be perturbed within a narrow neighborhood (e.g., ±20–30% relative adjustment).

Objective:

- Confirm signal performance does not depend on a single VWAP distance cutoff  
- Validate structural sensitivity to institutional positioning rather than precise numeric threshold  
- Identify stability plateaus rather than local maxima  

---

### C. Freshness Window Perturbation

The EMA cross freshness window (maximum bar age) will be perturbed ±1–2 bars.

Objective:

- Validate decay behavior of momentum edge  
- Ensure alpha does not vanish under small timing shifts  

---

### D. Stop & Risk Buffer Perturbation

The following will be tested within controlled neighborhoods:

- VWAP-anchored stop distance  
- Breakeven trigger threshold  
- EMA/HMA reversal buffer  

Objective:

- Confirm payoff profile stability  
- Ensure risk-return asymmetry persists across nearby configurations  
- Prevent overfitting to specific stop geometry  

---

### E. Volatility Threshold Perturbation

ATR percentile participation threshold will be perturbed across a reasonable range (e.g., ±5–10 percentile points).

Objective:

- Validate that signal strength is volatility-conditioned but not volatility-fragile  
- Confirm existence of a stability plateau rather than a sharp optimal point  

---

### F. Session Window Sensitivity

Statistically derived session windows will be expanded and contracted slightly to confirm:

- Signal density remains economically meaningful  
- Performance does not depend on narrow time slicing  

---

### Stability Criterion

For each perturbed dimension:

- Sharpe ratio should remain within an acceptable stability band  
- Trade count should not collapse  
- Drawdown profile should remain structurally similar  

The strategy is considered structurally robust only if performance remains stable across these independent perturbation dimensions.


## 10. Robustness & Anti-Overfitting Controls

Validation requires stability across:

- Time (static and rolling tests)
- Regimes (volatility segmentation)
- Cross-sectional universe members
- Parameter neighborhoods
- EMA pair perturbations
- Long vs short decomposition

Edge must remain statistically credible under multiple independent stress dimensions.

---

## 11. Evaluation Metrics

Performance will be assessed using:

- Sharpe ratio
- t-statistic of trade returns
- Maximum drawdown
- Win rate
- Trade count stability
- Regime-conditioned returns
- Return on deployed capital
- Long vs short contribution breakdown

Statistical credibility and structural stability are required for alpha validation.

---

## 12. Research Classification

This project represents:

> A correlation-defined, volatility-conditioned, regime-aware intraday alpha research framework governed by formal statistical parameter derivation and dual validation (static and adaptive) protocols.

The objective is not to produce an optimized historical backtest, but to determine whether a structurally defined signal family exhibits stable, statistically defensible alpha under disciplined, bias-aware research controls.
