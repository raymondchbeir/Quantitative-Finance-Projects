# PART I — MODEL A (CONTROL)
## Price-Derivative Momentum Framework

---

## 2. Structural Hypothesis — Model A

The core hypothesis is:

1. Persistent positioning relative to VWAP reflects intraday institutional directional bias.
2. Short-horizon EMA cross events represent momentum inflection points with decaying predictive value.
3. Volatility regimes materially affect signal reliability.
4. Structural alignment (VWAP + trend + momentum) reduces false positives.
5. Risk anchored to structural invalidation improves payoff convexity.
6. Intraday signal strength varies by time-of-day and must be statistically derived.

This defines the control signal family.

---

## 3. Structural (Non-Optimized) Definitions — Model A

The following define the modeling space and will **not** be statistically tuned:

- 5-minute bar resolution  
- EMA(9,20) as baseline short-horizon momentum pair  
- HMA(20) as structural trend filter  
- RSI(14) and default MACD parameters  
- Fixed notional position sizing  
- Initial capital baseline  

These define the hypothesis space.

---

## 4. Correlation-Defined Universe Construction — Model A

### 4.1 Anchor Asset

NVDA is designated as the anchor asset.

### 4.2 Universe Selection Methodology

For each candidate equity:

$ρ_i = Corr(r_i, r_NVDA)$

Where:

- $r_i$ = daily log returns of stock i  
- $r_NVDA$ = daily log returns of NVDA  
- Correlation window = 252 trading days  

Stocks are ranked by rolling correlation during a predefined universe formation period.

The top N correlated stocks form the trading universe.

The universe is frozen before the primary evaluation period to prevent look-ahead bias.

---

## 5. Statistically Governed Parameters (Applies to Model A)

The following variables will be empirically derived using controlled statistical analysis:

### Volatility Conditioning
- ATR lookback length
- ATR percentile threshold for participation

### Momentum Freshness
- EMA cross freshness window (maximum bar age)

### Structural Activation
- VWAP structural activation delay (bars after two-sided confirmation)
- VWAP proximity band at cross event

### Risk Management
- Initial stop-loss distance (VWAP-anchored)
- Breakeven trigger distance
- EMA/HMA structural reversal buffer

### Structural Trend Filters
- VWAP slope lookback window (longs)
- VWAP slope lookback window (shorts)

### Intraday Participation
- Statistically derived session windows based on time-of-day signal significance

Each parameter will be selected based on:

- Distributional analysis (MAE/MFE)
- Conditional forward return studies
- Stability plateaus across neighborhoods
- Regime robustness
- Cross-sectional consistency

No parameter will be selected solely on peak Sharpe.

---

## 6. Session Derivation Protocol — Model A

1. Identify all EMA cross events.
2. Compute forward returns at multiple horizons.
3. Group events by time-of-day buckets.
4. Compute:
   - Mean forward return
   - t-statistic
   - Cross-to-cross move distribution
5. Exclude time regimes where:
   - Forward returns are statistically insignificant, or
   - Cross-to-cross moves are economically negligible.

Session participation is data-driven.

# PART II — MODEL B (EXPERIMENTAL)
## Flow-Structural Microstructure Framework

---

## 7. Structural Hypothesis — Model B

Model B tests whether intraday alpha is better explained by:

1. Volume-weighted directional pressure rather than lagging price derivatives.
2. Auction-based structural acceptance (POC) rather than arbitrary smoothing filters.
3. Liquidity- and volatility-adaptive universe construction.
4. Institutional participation intensity via time-relative volume.
5. Explicit friction-aware execution modeling.

Model B challenges whether price momentum is the dominant short-horizon signal family.

---

## 8. Dynamic Universe Construction — Model B

### V1 Limitation
Correlation anchoring introduces:
- Idiosyncratic regime bias  
- Sector concentration drift  
- Structural macro dependence  

### V2 Implementation

Daily rebalanced universe using strictly lagged data:

1. Rank equities by 14-day ATR relative to price:
   ATR_Ratio_i = ATR_14_i / Price_i

2. Filter for:
   - Minimum average dollar liquidity
   - Above-average pre-market participation

3. Select Top 50 daily.

This ensures capital deployment where volatility and liquidity actually exist.

---

## 9. Signal Generation — Estimated Volume Delta (EVD)

Per-bar pressure proxy:

$EVD_t = V_t × ( 2 × (C_t - L_t) / (H_t - L_t) - 1 )$

Interpretation:

- Close near high → buyer dominance  
- Close near low → seller dominance  
- Volume-weighted aggressor pressure approximation  

To eliminate historical drag:

Flow_t = rolling sum of recent EVD  
or slope of EVD window.

Window length will be statistically derived and perturbation-tested.

---

## 10. Structural Trend Filter — Volume Profile & POC

Construct intraday volume profile using 1-minute or 5-minute data.

POC = price level with highest accepted traded volume during session.

Long condition requires:
- Price > VWAP
- Price > POC

Short condition requires:
- Price < VWAP
- Price < POC

POC resolution will be stability-tested across price bin sizes.

---

## 11. Institutional Participation Filter — Time-of-Day RVOL

Procedure:

1. Drop half-days and early closes.
2. Group bars by exact time-of-day.
3. Compute rolling median of last 20 valid same-time observations.

RVOL_t = V_t / Median(V_{t, last 20 valid same-time observations})

Require:
RVOL_t ≥ threshold (e.g., 2.0, perturbation-tested ± neighborhood).

This ensures structural activation only during institutional participation regimes.

# PART III — COMMON EXECUTION & VALIDATION

---

## 12. Execution Model (Shared)

- Signals computed at close of bar t.
- Entry executed at open of bar t+1.
- All exits triggered at close of bar t execute at open of bar t+1.
- No overnight exposure.
- One active position per asset.
- Multiple trades per day permitted.

PnL:

Long:
(Exit − Entry) × Qty

Short:
(Entry − Exit) × Qty

Performance evaluated separately:
- Long-only
- Short-only
- Combined

---

## 13. Transaction Cost Analysis (Applied to Both Models)

5 bps round-trip (2.5 bps per side).

Long Entry:
Actual_Entry = Theoretical_Open × (1 + 0.00025)

Long Exit:
Actual_Exit = Theoretical_Exit × (1 − 0.00025)

Equivalent mirrored logic for shorts.

Sensitivity analysis includes ±1–2 bps perturbation.

Alpha must survive cost-adjusted evaluation.
