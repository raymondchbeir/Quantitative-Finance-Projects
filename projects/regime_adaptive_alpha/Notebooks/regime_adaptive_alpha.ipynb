# Regime-Adaptive Intraday Alpha Model  
### Quantitative Research Overview

## 1. Research Objective

This project develops and evaluates a systematic intraday alpha model operating on 5-minute bars.  

The objective is to identify short-horizon directional inefficiencies using structural VWAP positioning, momentum confirmation, volatility conditioning, and regime-aware execution constraints.

The research focuses on:

- Structural price positioning relative to VWAP  
- Fresh momentum transitions (EMA cross dynamics)  
- Volatility regime filtering  
- Controlled intraday execution windows  
- Tight risk management anchored to market structure  

The goal is to determine whether these combined features produce statistically robust excess returns (alpha) after accounting for realistic trading constraints.

---

## 2. Research Hypothesis

We hypothesize that:

1. Persistent positioning relative to VWAP indicates intraday directional bias.
2. Fresh EMA cross events represent momentum inflection points with short-lived predictive value.
3. Volatility gating improves signal quality by filtering low-information regimes.
4. Conditioning entries on both structural alignment and momentum confirmation reduces false positives.
5. Anchoring risk to VWAP improves downside containment while preserving convex upside behavior.

If validated, this framework constitutes a regime-aware intraday alpha generation process.

---

## 3. Data Requirements

- 5-minute OHLCV bars
- Timezone-aware UTC index
- Precomputed indicators:
  - VWAP
  - EMA(9), EMA(20)
  - HMA(20)
  - RSI(14)
  - MACD, MACD Signal, MACD Histogram

Session segmentation is performed using U.S. Pacific Time for intraday window enforcement.

---

## 4. Signal Architecture

The alpha model integrates five components:

### A. Structural VWAP Bias

We detect two consecutive closes above or below VWAP to identify directional structural alignment.

Signal activation is delayed by 3 bars to avoid immediate breakout noise and allow confirmation.

---

### B. Momentum Freshness (EMA Cross Dynamics)

We identify EMA(9/20) cross events and track a freshness decay window (maximum 4 bars).

Only recent cross events are considered valid, reflecting the hypothesis that momentum edge decays quickly.

Cross events must occur near VWAP (±1%) to avoid late-stage extensions.

---

### C. Volatility Regime Filter

We compute ATR(5)/Close and require it to exceed the 40th percentile of its historical distribution.

This ensures participation only during sufficiently active volatility regimes.

---

### D. Trend & Momentum Confirmation

Additional alignment conditions include:

- HMA(20) directional filter
- VWAP slope confirmation
- RSI trend direction
- MACD confirmation (bearish confirmation required for shorts)

These conditions reduce structural countertrend entries.

---

### E. Intraday Execution Windows

Entries are restricted to:

- 07:00–09:30 PT (morning session)
- 11:00–12:50 PT (late session)

This avoids midday liquidity decay and end-of-day volatility distortions.

---

## 5. Execution Model

- Entry occurs at the current bar’s Open once all conditions align.
- Position sizing is fixed notional per trade.
- Only one position may be active at a time.

---

## 6. Risk Management Framework

### Initial Risk
Stops are anchored to VWAP at entry (0.5%).

### Breakeven Logic
Stops are advanced to entry once price moves +0.5% in favor.

### Dynamic Exit Triggers
- EMA/HMA structural reversal
- MACD inflection reversal
- End-of-session flattening
- End-of-data flattening

This framework prioritizes structural invalidation over arbitrary profit targets.
